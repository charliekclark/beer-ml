```python
metal_stats = {'bronze':1,
               'iron':5,
               'steel':10,
               'diamond':20}
```

```python
class VideoGameCharacter():
    def __init__(self, name):
        self.name_ = name

        #Stat Parameters
        self.attack_ = 1
        self.defense_ = 1
        self.hug_ = 1

        #Armor
        self.sword_ = 'bronze'
        self.shield_ = 'bronze'
    
    def found_sword(self, sword_type):
        self.sword_ = sword_type
        additional_attack = metal_stats[sword_type]
        self.attack_ = additional_attack

```

```python
guy = VideoGameCharacter(name = "Kyle")
print(guy.attack_, guy.sword_)
```

```python
guy.found_sword('iron')
```

```python
print(guy.sword_, guy.attack_)
```

```python
class FintechGrade():
    def __init__(self):
        self.assignments = {}
    
    def add_assignment(self, name, points, available):
        self.assignments[name] = {'points': points, 'available': available}
    
    def overall_grade(self):
        total_grade = 100 * sum(a['points'] for a in self.assignments.values()) / sum(a['available'] for a in self.assignments.values())
        print(f'Congrats! You have a {total_grade:.2f}% in Hunters bullshit class!')
```

```python
grades = FintechGrade()
```

```python
grades.add_assignment("bullshit", 50, 60)
grades.add_assignment("powerpoint formatting bullshit", 100, 101)
grades.add_assignment("ML 1", 67,75)
```

```python
grades.assignments
```

```python
grades.overall_grade()
```
